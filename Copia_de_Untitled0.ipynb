```python
def enriquecer_metadatos_con_ia(lista_de_chunks, modelo_llm):
    print(f"Iniciando el enriquecimiento de {len(lista_de_chunks)} trozos...")

    for i, chunk in enumerate(lista_de_chunks):
        prompt_clasificador = f"""
        Analiza el fragmento y responde ÚNICAMENTE con el nombre de la entidad:
        (PAMI, OSDE, OSER, Jerárquicos Salud, Galeno, CASA, OSECAC, OSPE, Swiss Medical).
        Si no es claro, responde 'General'.

        TEXTO: {chunk.page_content[:400]}
        """

        # Invocamos al modelo
        respuesta = modelo_llm.invoke(prompt_clasificador)

        # --- SOLUCIÓN AL ERROR ---
        # Si el contenido es una lista, extraemos el texto del primer elemento
        content = respuesta.content
        if isinstance(content, list):
            # Buscamos la parte que contiene el texto
            entidad = content[0].get('text', str(content[0])) if isinstance(content[0], dict) else str(content[0])
        else:
            entidad = str(content)

        chunk.metadata["entidad"] = entidad.strip()
        # -------------------------

        if (i + 1) % 15 == 0:
            print(f"Procesados {i + 1} de {len(lista_de_chunks)} trozos...")

    print("Enriquecimiento completado con éxito.")
    return lista_de_chunks
```

```python
import time
from langchain_community.vectorstores import FAISS

# 1. Configuramos el modelo (como ya lo tenías)
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")

# 2. Función de Analista: Crear la base de datos por lotes (Batching)
# Esto evita el error 429 y garantiza que no perdamos datos
def crear_vectorstore_por_lotes(documentos, modelo_embeddings, lote_size=20):
    vector_db = None

    # Dividimos los 78 trozos en grupos de 20
    for i in range(0, len(documentos), lote_size):
        lote = documentos[i : i + lote_size]
        print(f"Procesando lote de trozos {i} a {i + len(lote)}...")

        if vector_db is None:
            # El primer lote crea la base de datos
            vector_db = FAISS.from_documents(lote, modelo_embeddings)
        else:
            # Los siguientes lotes se agregan a la base ya existente
            vector_db.add_documents(lote)

        # Pausa de seguridad de 10 segundos entre lotes para respetar el Rate Limit
        print("Esperando 10 segundos para no saturar la API...")
        time.sleep(10)

    return vector_db

# Ejecutamos la función con tus 78 chunks
vectorstore = crear_vectorstore_por_lotes(chunks, embeddings)

print("\n¡Base de datos de normativas creada exitosamente!")
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

# Configuramos el buscador
retriever = vectorstore.as_retriever(search_kwargs={"k": 4})

# Formateador para mostrar Fuente y Entidad en el contexto
def format_docs_with_metadata(docs):
    return "\n\n".join([f"--- ENTIDAD: {d.metadata.get('entidad')} (Pág: {d.metadata.get('page')}) ---\n{d.page_content}" for d in docs])

template = """
Eres un auditor de farmacia estricto. Tu objetivo es realizar un triage burocrático para evitar débitos.
Utiliza solo el contexto provisto. Si no sabes, di que no hay información suficiente.

CONTEXTO:
{context}

PREGUNTA:
{question}

RESPUESTA:
"""
prompt = ChatPromptTemplate.from_template(template)

rag_chain = (
    {"context": retriever | format_docs_with_metadata, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

print("Pipeline RAG de alta fidelidad completado.")
```
